```python
import pandas as pd
```

```python
df = pd.read_csv("D:/spam.csv", encoding='windows-1252')
df
```

```python
df.columns
```

```python
df = df.drop(['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'], axis=1)
```

```python
df
```

```python


# Renaming specific columns
df = df.rename(columns={'v1': 'label', 'v2': 'sms_message'})
df
```

```python
df.label.unique()
```

```python
 df["label"]=df.label.map({"ham":0,"spam":1})
```

```python
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
count_vector=CountVectorizer()
```

```python
count_vector
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(df["sms_message"],df["label"],random_state=1)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
df["sms_message"].shape
```

```python
train=count_vector.fit_transform(x_train)
```

```python
test=count_vector.transform(x_test)
```

```python
train
```

```python
train.toarray()
```

```python
count_vector.get_feature_names_out()
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
naive_bayes=MultinomialNB()
```

```python
naive_bayes.fit(train,y_train)
```

```python
predictions=naive_bayes.predict(test)
```

```python
predictions
```

```python
predlist=list(zip(predictions,y_test))
```

```python
predlist
```

```python
from sklearn.metrics import accuracy_score,precision_score
```

```python
score=accuracy_score(y_test,predictions)
score
```
